```python
import qiskit
from qiskit import *
```

```python
from qiskit.providers.aer.noise import NoiseModel
from qiskit.providers.aer.noise import pauli_error, depolarizing_error
```

```python
def get_noise(p):
    error_meas=pauli_error([('X',p),('I',1-p)])
    noise_model=NoiseModel()
    noise_model.add_all_qubit_quantum_error(error_meas,"measure")
    
    return noise_model
```

```python
noise_model=get_noise(0.01)
```

```python
for state in ['00','01','10','11']:
    qc = QuantumCircuit(2,2)
    if state[0]=='1':
        qc.x(1)
    if state[1]=='1':
        qc.x(0)  
    qc.measure(qc.qregs[0],qc.cregs[0])
    print(state+' becomes',
          execute(qc,Aer.get_backend('qasm_simulator'),noise_model=noise_model,shots=10000).result().get_counts())
```

#### Here we find that the correct output is certainly the most dominant. Ones that differ on only a single bit (such as '01', '10' in the case that the correct output is '00' or '11'), occur around  1%  of the time. Those than differ on two bits occur only a handful of times in 10000 samples, if at all.

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)  
qc.measure(qc.qregs[0],qc.cregs[0])
print(execute(qc,Aer.get_backend('qasm_simulator'),noise_model=noise_model,shots=10000).result().get_counts())
```

#### Here '01' and '10' occur for around  1%  of all samples. We know from our analysis of the basis states that such a result can be expected when these outcomes should in fact never occur, but instead the result should be something that differs from them by only one bit: '00' or '11'. Now we just need to find a way to perform the mitigation algorithmically rather than manually. We will do this by describing the random process using matrices.

```python
from qiskit.ignis.mitigation.measurement import (complete_meas_cal,CompleteMeasFitter)
```

```python
qr = qiskit.QuantumRegister(3)
meas_calibs, state_labels = complete_meas_cal(qr=qr, circlabel='mcal')
```

```python
for circuit in meas_calibs:
    print("Circuit ", circuit.name)
    print(circuit)
```

```python
backend = qiskit.Aer.get_backend('qasm_simulator')
job = qiskit.execute(meas_calibs, backend=backend, shots=1000,noise_model=noise_model)
cal_results = job.result()
```

```python
meas_fitter = CompleteMeasFitter(cal_results, state_labels, circlabel='mcal')
corr_matrix=meas_fitter.cal_matrix
print(corr_matrix)
```

### Let's see how well we can mitigate for this noise. Again, let's use the state  (|000⟩+|001⟩+|110⟩+|111⟩)/2  for our test.

```python
from qiskit.visualization import plot_histogram
```

# This is the expected histogram

```python
circuit=QuantumCircuit(3,3)
circuit.h(0)
circuit.cx(0,1)
circuit.cx(0,2)
circuit.h(2)
circuit.measure(0,0)
circuit.measure(1,1)
circuit.measure(2,2)
result=execute(circuit,backend=Aer.get_backend("qasm_simulator"),shots=2000).result()
counts=result.get_counts()
plot_histogram(counts)
```

```python
result=execute(circuit,backend=Aer.get_backend("qasm_simulator"),shots=2000, noise_model=noise_model).result()
noisy_counts=result.get_counts()
plot_histogram(noisy_counts)
```

```python
meas_filter = meas_fitter.filter
# Results with mitigation
mitigated_results = meas_filter.apply(result)
mitigated_counts = mitigated_results.get_counts()
```

```python
%config InlineBackend.figure_format = 'svg' # Makes the images look nice
plot_histogram([noisy_counts, mitigated_counts], legend=['Noisy', 'Mitigated'])
```

# Nice. Now on a real quantum computer

```python
from qiskit import IBMQ
IBMQ.save_account("",overwrite=True)
IBMQ.load_account()
```

```python
provider = IBMQ.get_provider(hub='ibm-q')
provider.backends()
```

```python
from qiskit.providers.ibmq import least_busy
backend = least_busy(provider.backends(filters=lambda b: b.configuration().n_qubits >= 3 and
                                   not b.configuration().simulator and b.status().operational==True))
```

```python
qr = qiskit.QuantumRegister(3)
meas_calibs, state_labels = complete_meas_cal(qr=qr)
```

```python
job = execute(meas_calibs, backend=backend, shots=8192)
```

```python
from qiskit.tools.monitor import job_monitor
job_monitor(job)
```

```python
cal_results= job.result()
meas_fitter=CompleteMeasFitter(cal_results,state_labels)
corr_matrix=meas_fitter.cal_matrix
print(corr_matrix)
```

```python
#We will use the same circuit we used before
circuit.draw(output="mpl")
```

```python
job_circuit= execute(circuit, backend=backend, shots=8192)
```

```python
job_monitor(job_circuit)
```

```python
circuit_result=job_circuit.result()
noisy_counts=circuit_result.get_counts()
```

```python
meas_fitter = meas_fitter.filter
# Results with mitigation
mitigated_results = meas_filter.apply(circuit_result)
mitigated_counts = mitigated_results.get_counts()
```

```python
plot_histogram([noisy_counts, mitigated_counts], legend=['Noisy', 'Mitigated'])
```
