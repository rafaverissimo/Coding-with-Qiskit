```python
import qiskit 
import numpy as np
from numpy import pi
from qiskit import *
from qiskit.providers.ibmq import least_busy
from qiskit.tools.monitor import job_monitor
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
def qft_circuit(circuit,n):
    s=n
    while n>0:
        n-=1
        circuit.h(n)
        for qubit in range(n):
            circuit.cu1(pi/2**(n-qubit),qubit,n) 
    
    for qubit in range(s//2):
        circuit.swap(qubit, s-qubit-1)
    circuit.barrier()
    return circuit
```

```python
qc=QuantumCircuit(4,4)
qc=qft_circuit(qc,4)
```

```python
qc.draw("mpl")
```

### Let's use this state as initial state to test the inverse QFT

```python
number=10
qc = QuantumCircuit(4,4)
for qubit in range(4):
    qc.h(qubit)
qc.u1(number*pi/8,0)
qc.u1(number*pi/4,1)
qc.u1(number*pi/2,2)
qc.u1(number*pi,3)
qc.draw('mpl')
```

```python
statevector = execute(qc, backend=backend).result().get_statevector()
plot_bloch_multivector(statevector)
```

```python
def inverse_qft(circuit, n):
    """Does the inverse QFT on the first n qubits in circuit"""
    # First we create a QFT circuit of the correct size:
    qft_circ = qft_circuit(QuantumCircuit(n), n)
    # Then we take the inverse of this circuit
    invqft_circ = qft_circ.inverse()
    # And add it to the first n qubits in our existing circuit
    circuit.append(invqft_circ, circuit.qubits[:n])
    return circuit.decompose() 
```

```python
qc=inverse_qft(qc,4)
qc.barrier()
qc.measure([0,1,2,3],[0,1,2,3])
qc.draw("mpl")
```

```python
# Load our saved IBMQ accounts and get the least busy backend device with less than or equal to nqubits
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q')
backend = least_busy(provider.backends(filters=lambda x: x.configuration().n_qubits >= 4
                                       and not x.configuration().simulator 
                                       and x.status().operational==True))
print("least busy backend: ", backend)
```

```python
job = execute(qc, backend=backend, shots = 2048)
job_monitor(job)
```

```python
counts = job.result().get_counts()
plot_histogram(counts)
```
