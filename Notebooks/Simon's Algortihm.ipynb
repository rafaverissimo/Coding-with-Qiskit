```python
# Calculate the dot product of the results
def sdotz(a, b):
    accum = 0
    for i in range(len(a)):
        accum += int(a[i]) * int(b[i])
    return (accum % 2)
```

```python
# Measures of the first n=len(s) qubits
def meas(s,result):
    # Categorize measurements by input register values
    counts=result.get_counts()
    answer = {}
    for meas in counts.keys():
        qm=meas[len(s):]
        if qm in answer:
            answer[qm]+= counts[meas]
        else:
            answer[qm]= counts[meas]
            
    print('s, z, s.z (mod 2)')
    for z in answer:
        print( '{}, {}, {}.{}={}'.format(s, z, s,z,sdotz(s,z)) )
```

```python
meas(s, result)
```

# Simon's general algorithm 

```python
import random
from random import randint
```

```python
def simon(s):
    n=len(s)
    qc=QuantumCircuit(2*n,2*n)
    for i in range(n):
        qc.h(i)
    qc.barrier()
    
    #Oracle
    for i in range(n):
        qc.cx(i,n+i)
    
    if int(s)!=0:
        for i in range(n):
            qc.cx(n-1,n+i)
            r=randint(0,1)
            if r==1:
                qc.x(n+i)
    qc.barrier()
    for i in range(n):
        qc.h(i)
    for i in range(n):
        qc.measure(i,i)
        qc.measure(n+i,n+i)
    
   
    return qc 
```

```python
s="010"
simon_circ=simon(s)
```

```python
simon_circ.draw("mpl")
```

```python
from qiskit.providers.ibmq import least_busy
from qiskit.tools.monitor import job_monitor
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q')
provider.backends()
backend = least_busy(provider.backends(filters=lambda x: x.configuration().n_qubits >= 6 and 
                                   not x.configuration().simulator and x.status().operational==True))
print("least busy backend: ", backend)
```

```python
job=execute(simon_circ,backend=backend,shots=1024)
job_monitor(job)
```

```python
result=job.result()
meas(s,result)
```
